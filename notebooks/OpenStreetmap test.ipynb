# OpenStreetmap
Example to extract waterways from OpenStreetmap

Requires:

    pip install overpass

```python
import overpass
```

```python
api = overpass.API()
```

From google earth I gathered coordinates for a bounding box focussing on the Amsterdam centre (lower left coordinate, upper right coordinate). 

I next got all 'way' features in this bounding box and list all geometry features where type is LineString. A further filtering is needed and likely possible. There may be tools available to turn this information directly into a networkx graph. 

```python
data = api.get('way(52.36309012572587,4.890326718121118,52.36590601699889,4.898316757037793);(._;>;)', verbosity='geom')
[f for f in data.features  if f.geometry['type'] == "LineString"]
```
